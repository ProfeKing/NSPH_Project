```python
# Francesca e Nicola
'''nuova tabella di colesterolo con valori pari a 1 2 e 3'''
def categorise(row):  
    if row['chol'] <= 200:
        return 1
    elif row['chol'] > 200 and row['chol'] <= 239:
        return 2
    elif row['chol'] >= 240:
        return 3
    
heart['chol_lvl'] = heart.apply(lambda row: categorise(row), axis=1)

desiderable_level = heart['chol_lvl'] == 1
borderline_level = heart['chol_lvl'] == 2
high_level = heart['chol_lvl'] == 3

#print(len(heart))
print("numero di pazienti con un /desiderable level/ di colesterolo:",len(desiderable_level))
print("numero di pazienti con un /borderline level/ di colesterolo:",len(borderline_level))
print("numero di pazienti con un /high level/ di colesterolo:",len(high_level))
print("totale pazienti",len(desiderable_level) + len(borderline_level) + len(high_level))


healthy = heart[heart["target"]==0]
sick = heart[heart["target"]==1]

print("numero di pazienti sani:",len(healthy))
print("numero di pazienti malati:",len(sick))
print("totale pazienti",len(healthy)+len(sick))


'''PAZIENTI CON MALATTIE CARDIACHE IN RELAZIONE AL LIVELLO DI COLESTEROLO'''

des_sick = heart[(heart['chol_lvl'] == 1) & (heart['target'] == 1)] #desiderable level di colesterolo nei soggetti con malattie cardiache
bord_sick = heart[(heart['chol_lvl'] == 2) & (heart['target']==1)] #borderline level di colesterolo nei soggetti con malattie cardiache
high_sick = heart[(heart['chol_lvl'] == 3) & (heart['target'] == 1)] #high level di colesterolo nei soggetti con malattie cardiache

print('pazienti con malattie cardiache e un /desiderable level/ di colesterolo:',len(des_sick),"-",round(len(des_sick)/len(sick)*100,2),"%","sul totale dei pazienti malati")
print('pazienti con malattie cardiache e un /borderline level/ di colesterolo:',len(bord_sick),"-",round(len(bord_sick)/len(sick)*100,2),"%","sul totale dei pazienti malati")
print('pazienti con malattie cardiache e un /high level/ di colesterolo:',len(high_sick),"-",round(len(high_sick)/len(sick)*100,2),"%","sul totale dei pazienti malati")


'''PAZIENTI SENZA MALATTIE CARDIACHE PERVENUTE IN RELAZIONE AL LIVELLO DI COLESTEROLO'''

des_healthy = heart[(heart['chol_lvl'] == 1) & (heart['target'] == 0)] #desiderable level di colesterolo nei soggetti senza malattie cardiache
bord_healthy = heart[(heart['chol_lvl'] == 2) & (heart['target'] == 0)] #borderline level di colesterolo nei soggetti senza malattie cardiache
high_healthy = heart[(heart['chol_lvl'] == 3) & (heart['target'] == 0)] #high level di colesterolo nei soggetti senza malattie cardiache

print('pazienti senza malattie pervenute cardiache e un /desiderable level/ di colesterolo:',len(des_healthy),"-",round(len(des_healthy)/len(healthy)*100,2),"%","sul totale dei pazienti sani")
print('pazienti senza malattie pervenute cardiache e un /borderline level/ di colesterolo:',len(bord_healthy),"-",round(len(bord_healthy)/len(healthy)*100,2),"%","sul totale dei pazienti sani")
print('pazienti senza malattie pervenute cardiache e un /high level/ di colesterolo:',len(high_healthy),"-",round(len(high_healthy)/len(healthy)*100,2),"%","sul totale dei pazienti sani")


nmale = len(heart[heart["sex"]==1])
nfemale = len(heart[heart["sex"]==0])

print("nuemro di pazienti uomo",nmale)
print("nuemro di pazienti donna",nfemale)
print("pazienti totali",nmale+nfemale)

'''PAZIENTI MASCHI IN RELAZIONE AL LIVELLO DI COLESTEROLO'''

des_male= heart[(heart['chol_lvl'] == 1) & (heart['sex'] == 1)] #desiderable level di colesterolo nei soggetti maschi
bord_male = heart[(heart['chol_lvl'] == 2) & (heart['sex'] == 1)] #borderline level di colesterolo nei soggetti maschi
high_male = heart[(heart['chol_lvl'] == 3) & (heart['sex'] == 1)] #high level di colesterolo nei soggetti maschi

print('pazienti maschi con un /desiderable level/ di colesterolo:',len(des_male),"-",round(len(des_male)/nmale*100,2),"%","sul totale dei pazienti maschi")
print('pazienti maschi con un /borderline level/ di colesterolo:',len(bord_male),"-",round(len(bord_male)/nmale*100,2),"%","sul totale dei pazienti maschi")
print('pazienti maschi con un /high level/ di colesterolo:',len(high_male),"-",round(len(high_male)/nmale*100,2),"%","sul totale dei pazienti maschi")


'''PAZIENTI DONNE IN RELAZIONE AL LIVELLO DI COLESTEROLO'''

des_female= heart[(heart['chol_lvl'] == 1) & (heart['sex'] == 0)] #desiderable level di colesterolo nelle donne
bord_female = heart[(heart['chol_lvl'] == 2) & (heart['sex'] == 0)] #borderline level di colesterolo nelle donne
high_female = heart[(heart['chol_lvl'] == 3) & (heart['sex'] == 0)] #high level di colesterolo nelle donne

print('pazienti donne con un /desiderable level/ di colesterolo:',len(des_female),"-",round(len(des_female)/nfemale*100,2),"%","sul totale dei pazienti donne")
print('pazienti donne con un /borderline level/ di colesterolo:',len(bord_female),"-",round(len(bord_female)/nfemale*100,2),"%","sul totale dei pazienti donne")
print('pazienti donne con un /high level/ di colesterolo:',len(high_female),"-",round(len(high_female)/nfemale*100,2),"%","sul totale dei pazienti donne")


'''PAZIENTI UOMINI IN RELAZIONE AL LIVELLO DI COLESTEROLO E AL LIVELLO DI ZUCCHERO NEL SANGUE'''
des_male_NoDiabetes= heart[(heart['chol_lvl'] == 1) & (heart['sex'] == 1) & (heart['fbs']==0)]
des_male_Diabetes= heart[(heart['chol_lvl'] == 1) & (heart['sex'] == 1) & (heart['fbs']==1)]

print('pazienti maschi con un /desiderable level/ di colesterolo e con un /normale livello di zuccheri/ nel sangue:',len(des_male_NoDiabetes),"-",round(len(des_male_NoDiabetes)/nmale*100,2),"%", "sul totale dei maschi")
print('pazienti maschi con un /desiderable level/ di colesterolo e con /diabete/:',len(des_male_Diabetes),"-",round(len(des_male_Diabetes)/nmale*100,2),"%")

bord_male_NoDiabetes= heart[(heart['chol_lvl']==2) & (heart['chol']<=239) & (heart['sex']==1) & (heart['fbs']==0)]
bord_male_Diabetes= heart[(heart['chol_lvl']==2) & (heart['chol']<=239) & (heart['sex']==1) & (heart['fbs']==1)]

print('pazienti maschi con un /borderline level/ di colesterolo e con un /normale livello di zuccheri/ nel sangue:',len(bord_male_NoDiabetes),"-",round(len(bord_male_NoDiabetes)/nmale*100,2),"%", "sul totale dei maschi")
print('pazienti maschi con un /borderline level/ di colesterolo e con /diabete/:',len(bord_male_Diabetes),"-",round(len(bord_male_Diabetes)/nmale*100,2),"%")

high_male_NoDiabetes= heart[(heart['chol_lvl']==3) & (heart['sex'] == 1) & (heart['fbs']==0)]
high_male_Diabetes= heart[(heart['chol_lvl'] ==3) & (heart['sex'] == 1) & (heart['fbs']==1)]

print('pazienti maschi con un /high level/ di colesterolo e con un /normale livello di zuccheri/ nel sangue:',len(high_male_NoDiabetes),"-",round(len(high_male_NoDiabetes)/nmale*100,2),"%", "sul totale dei maschi")
print('pazienti maschi con un /high level/ di colesterolo e con /diabete/:',len(high_male_Diabetes),"-",round(len(high_male_Diabetes)/nmale*100,2),"%")


'''PAZIENTI DONNE IN RELAZIONE AL LIVELLO DI COLESTEROLO E AL LIVELLO DI ZUCCHERO NEL SANGUE'''
des_female_NoDiabetes= heart[(heart['chol_lvl'] ==1) & (heart['sex'] == 0) & (heart['fbs']==0)]
des_female_Diabetes= heart[(heart['chol_lvl'] ==1) & (heart['sex'] == 0) & (heart['fbs']==1)]

print('pazienti donne con un /desiderable level/ di colesterolo e con un /normale livello di zuccheri/ nel sangue:',len(des_female_NoDiabetes),"-",round(len(des_female_NoDiabetes)/nfemale*100,2),"%")
print('pazienti donne con un /desiderable level/ di colesterolo e con /diabete/:',len(des_female_Diabetes),"-",round(len(des_female_Diabetes)/nfemale*100,2),"%")

bord_female_NoDiabetes= heart[(heart['chol_lvl']==2) & (heart['chol']<=239) & (heart['sex']==0) & (heart['fbs']==0)]
bord_female_Diabetes= heart[(heart['chol_lvl']==2) & (heart['chol']<=239) & (heart['sex']==0) & (heart['fbs']==1)]

print('pazienti donne con un /borderline level/ di colesterolo e con un /normale livello di zuccheri/ nel sangue:',len(bord_female_NoDiabetes),"-",round(len(bord_female_NoDiabetes)/nfemale*100,2),"%")
print('pazienti donne con un /borderline level/ di colesterolo e con /diabete/:',len(bord_female_Diabetes),"-",round(len(bord_female_Diabetes)/nfemale*100,2),"%")

high_female_NoDiabetes= heart[(heart['chol_lvl'] ==3) & (heart['sex'] == 0) & (heart['fbs']==0)]
high_female_Diabetes= heart[(heart['chol_lvl'] ==3) & (heart['sex'] == 0) & (heart['fbs']==1)]

print('pazienti donne con un /high level/ di colesterolo e con un /normale livello di zuccheri/ nel sangue:',len(high_female_NoDiabetes),"-",round(len(high_female_NoDiabetes)/nfemale*100,2),"%")
print('pazienti donne con un /high level/ di colesterolo e con /diabete/:',len(high_female_Diabetes),"-",round(len(high_female_Diabetes)/nfemale*100,2),"%")
```
